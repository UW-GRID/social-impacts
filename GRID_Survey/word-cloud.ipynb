```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import pandas as pd
import sys
from collections import defaultdict
```

```python
csv_file = 'INITIA_SURVEY_ANALYSIS_RAW.csv'
df =pd.read_csv(csv_file)

#master_of_none = df.loc[df.notnull()]
```

```python
column_header1 = 'Household occupations - Coded'
column_header2 = 'Other common occupation - Coded'
subset1 = df[column_header1]
subset2 = df[column_header2]
dataset = str(df[[column_header1, column_header2]])
```

```python
wordcloud = WordCloud(width=500, height=500, colormap="hsv", background_color="black", margin=0).generate(dataset)
```

```python
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.margins(x=0, y=0)
plt.show
```
